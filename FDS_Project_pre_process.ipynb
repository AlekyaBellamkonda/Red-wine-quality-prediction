```
dataO = df.drop('quality',axis = 1)
features = dataO.columns
for feature in features :
  Lower_value= df[feature].mean() - 3*df[feature].std()
  Upper_value= df[feature].mean() + 3*df[feature].std()
  print(feature)
  print(Lower_value)
  print(Upper_value)
  for i in range(df.shape[0]) :
      if (df[feature][i] < Lower_value) | (df[feature][i] > Upper_value):
          df[feature][i] = df[feature].mean()
                              
```

```
new_dataframe1 = dataframe.replace({'quality' : [1,2,3,4,5] },0)
new_datafarme = new_dataframe1.replace({'quality' : [6,7,8,9,10]},1)
new_datafarme
```

```
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
discriptive_features = new_datafarme.drop('quality',axis =1)
discriptive_DF = pd.DataFrame(data = discriptive_features)
numerical = discriptive_features.columns
normalized_DF = pd.DataFrame(data = discriptive_DF)
normalized_DF[numerical] = scaler.fit_transform(discriptive_DF[numerical])
normalized_DF['quality']=new_datafarme.quality
display(normalized_DF.head(n = 1599))
normalized_DF['quality'].value_counts().plot(kind='bar')
```

```
for feature in features :
    print(feature)
    plt.hist(normalized_DF[feature],bins = 20,align='right')
    # plt.ylim((0, 10000))
    # plt.yticks([0, 500, 1000, 1500, 2000])
    plt.xticks( rotation='vertical')
    plt.show()
    #sns.pairplot(normalized_DF, hue='quality')
```

```
correlations = normalized_DF.corr()
correlations
plt.figure(figsize=(12,10))
sns.heatmap(correlations, annot=True, cmap=plt.cm.Reds)
```

```
display(correlations)
```

```
cor_target = abs(correlations["quality"])
cor_target
```

```
relevant_features = cor_target[cor_target>0.05]
print("relevant features")
display(relevant_features)
irrelevant_features=cor_target[cor_target<0.05]
print("irrelevant features")
display(irrelevant_features)
```
