```
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score,f1_score
from sklearn.metrics import classification_report
from sklearn import metrics
```

loading the preprocessed data set

```
dataframe=pd.read_csv('normalized_data.csv')
dataframe
```

Here we are copying the dataset int a dataframe called data


```
data = dataframe.copy()
data = data.drop('Unnamed: 0',axis = 1)
x = data.drop('quality',axis=1)
y = data['quality']
data
```

Here we are spliting the data inorder to apply the Decision Tree classifer


```
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.5)
clf = DecisionTreeClassifier(max_depth=6)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
Training_score = clf.score(X_train,y_train)
Testing_score = clf.score(X_test,y_test)
print('training score is {} testing score is {}'.format(Training_score,Testing_score))
```

```
pred = clf.predict(X_test)
print("Confusion Matrix:" "\n",metrics.confusion_matrix(y_test, clf.predict(X_test)))
print()
print("Classification_Report:" "\n", metrics.classification_report(y_test, clf.predict(X_test))) 
print()
Accuracy = accuracy_score(y_test,pred)
print( Accuracy)
```

Here we are applying the Gaussian Navie Bayes model and Displayed the classifiaction Report

```
clf2 = GaussianNB()
clf2.fit(X_train,y_train)
pred2 = clf2.predict(X_test)
Accuracy2 = accuracy_score(y_test,pred2)
print("the accuracy of the model is:", Accuracy2)
print("Confusion Matrix:" "\n",metrics.confusion_matrix(y_test, clf2.predict(X_test)))
print()
print("Classification_Report:" "\n", metrics.classification_report(y_test, clf2.predict(X_test))) 
print()
Training_score = clf2.score(X_train,y_train)
Testing_score = clf2.score(X_test,y_test)
print('training score is {} testing score is {}'.format(Training_score,Testing_score))
```

Here we use different "k" values and try to analyze the proposed model and then we perform comaprisons and then at last based on the training and testing score we find the best "K" value, which oin our case would be k=7

```
knnscores = pd.DataFrame(columns= ['K','Training score','Testing score'])
Training_score = []
Testing_score = []
k =[]
for i in range(1,25) :
    knnclf = KNeighborsClassifier(n_neighbors=i, p=1, metric='minkowski')
    knnclf.fit(X_train,y_train)
    Training_score.append(knnclf.score(X_train,y_train))
    Testing_score.append(knnclf.score(X_test,y_test))
    k.append(i)
knnscores['K'] = k
knnscores['Training score'] = Training_score
knnscores['Testing score'] = Testing_score
#print(knnscores)
print("Confusion Matrix:" "\n",metrics.confusion_matrix(y_test, knnclf.predict(X_test)))
print()
print("Classification_Report:" "\n", metrics.classification_report(y_test, knnclf.predict(X_test))) 
print()
knnscores
```
